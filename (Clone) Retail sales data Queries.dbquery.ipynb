```sql
select * from retail_sales_dataset limit 100;

--Display only the customer Id and product category columns

SELECT `Customer ID`,
       `Product Category`
       FROM retail_sales_dataset;


--Display all electronic sales transactions

SELECT *
FROM retail_sales_dataset
WHERE `Product Category`='Electronics';

--Display customer who are older than 40
SELECT *
FROM retail_sales_dataset
WHERE AGE>40;


--Display Beauty product sales where the total amount is greater than 100 

SELECT * 
FROM retail_sales_dataset
WHERE `Product Category`='Beauty' AND `Total Amount` > 100;

--Display the Customers who's age is between 30 and 65

SELECT * 
FROM retail_sales_dataset
WHERE Age BETWEEN 30 AND 65 LIMIT 10;




```
